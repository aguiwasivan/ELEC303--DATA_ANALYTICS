```python
car['Accident_Severity'] = car ['Accident_Severity'].astype("category")
car['Accident Date'] = car ['Accident Date'].astype("category")
car['Light_Conditions'] = car ['Light_Conditions'].astype("category")
car['District Area'] = car ['District Area'].astype("category")
car['Road_Surface_Conditions'] = car ['Road_Surface_Conditions'].astype("category")
car['Road_Type'] = car ['Road_Type'].astype("category")
car['Urban_or_Rural_Area'] = car ['Urban_or_Rural_Area'].astype("category")
car['Weather_Conditions'] = car ['Weather_Conditions'].astype("category")
car['Vehicle_Type'] = car ['Vehicle_Type'].astype("category")
```

```python
car['Latitude'] = car['Latitude'].fillna(car['Latitude'].mode()[0])
car['Longitude'] = car['Longitude'].fillna(car['Longitude'].mode()[0])
car['Road_Surface_Conditions'] = car['Road_Surface_Conditions'].fillna(car['Road_Surface_Conditions'].mode()[0])
car['Road_Type'] = car['Road_Type'].fillna(car['Road_Type'].mode()[0])
car['Longitude'] = car['Longitude'].fillna(car['Longitude'].mode()[0])
car['Urban_or_Rural_Area'] = car['Urban_or_Rural_Area '].fillna(car['Urban_or_Rural_Area '].mode()[0])
car['Weather_Conditions'] = car['Weather_Conditions'].fillna(car['Weather_Conditions'].mode()[0])
```
